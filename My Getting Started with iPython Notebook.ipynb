## Conditional statements in Python

```python
if i == 1 and f > 4:
    print 'The value of i is 1 and f is greater than 4.'
elif i > 4 or f > 4:
    print "i & f are both greater than 4."
else:
    print "Both i & f are less than or equal to 4."
```
